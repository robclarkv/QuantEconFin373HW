```python
import pandas as pd
import numpy as np
```

```python
url = "https://datascience.quantecon.org/assets/data/bball.csv"
bball = pd.read_csv(url)
bball.info()
```

```python
bball_wide = bball.drop("TeamName", axis=1).set_index(["Year", "Player", "Team"]).stack().unstack(level=[1, 3, 2]).sort_index(axis=1)
bball_wide
```

Can you think of a reason to ever use pivot rather than pivot_table? Write your thoughts.

Create a pivot table with column Player as the index, TeamName as the columns, and [Rebound, Assist] as the values. What happens when you use aggfunc=[np.max, np.min, len]? Describe how Python produced each of the values in the resultant pivot table.

### Answer:
One reason to use pivot instead of pivot_table is when we don't need to perform any aggregation or if we want to use a specific aggregation function that is not available in pivot_table. pivot simply reshapes the data from long to wide format based on the specified columns, while pivot_table performs aggregation on the values in the table based on the specified columns.



```python
 # To create a pivot table with the specified index, 
 # columns, and values, we can use the following code:
pvt = bball.pivot_table(index='Player', columns='TeamName', values=['Rebound', 'Assist'])
pvt
```

```python
# To use aggfunc=[np.max, np.min, len] on the values in the pivot table,
# we can modify the code as follows:
pvt = bball.pivot_table(index='Player', columns='TeamName', values=['Rebound', 'Assist'], aggfunc=[np.max, np.min, len])
pvt
```

This will produce a pivot table with the maximum, minimum, and count of the "Rebound" and "Assist" values for each combination of "Player" and "TeamName". The resulting DataFrame will have a multi-level column index with the following levels: the function names ("amax", "amin", "len"), the variable names ("Assist", "Rebound"), and the TeamName values. The amax function computes the maximum value, amin computes the minimum value, and len computes the number of non-null values in each group.
